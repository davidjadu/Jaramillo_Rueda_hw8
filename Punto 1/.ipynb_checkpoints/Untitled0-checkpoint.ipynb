#Análisis de Imágenes de vesículas utilizando métodos de MCMC y Transformada de Fourier
En este Trabajo se quiere estudiar los bordes de algunas imágenese de vesículas y de qué forma es el borde estás, para esto se seguirá el siguiente procedimiento:

##Filtro imágen
En primer Lugar se hace un filtro usando la transformada de fourier para permitir que se vean los bordes de la imágen con más claridad. Para más información ver:
http://nbviewer.ipython.org/github/davidjadu/Jaramillo_Solano_hw5/blob/master/FiltrandoImagen/leer_imagenes_tiff.ipynb

##Normalización de la Imágen
En este punto se normaliza la imágen para que todos los valores queden entre 0 y 1. Ya que es más cómodo para trabajar. Además se eliminan algunos puntos interiores y exteriores a los bordes, para tener una imágen más limpia.

##Modelamiento
Para modelar la forma circular de los bordes se hará un MCMC. Con el siguiente modelo para el radio en función del ángulo theta.

$$
r(\theta) = r_0 + r_1\sin(m\theta+\varphi)
$$

Además ya que tampoco se conoce el centro se van a utilizar las variables $a$ y $b$ que representan el centro de la figura. Para todas las variables se va a utilizar una distribución gaussiana como la vista en clase, excepto para $m$ ya que éste debe ser un entero y por tanto se utilizará un distribución de Poisson. Note que ya que cuando en una distribución de Poisson la media se va a 0, el único valor posible es 0, entonces se hizo una pequeña correción y cuando $m=0$ se toma una distribución de poisson con paramétro $\lambda = 0.5$ ($\mathcal{P}(0.5)$).

```python
%pylab inline
import matplotlib.pyplot as plt
from scipy.fftpack import fft, fftfreq
```

#Imágen 1
##Filtro imágen

```python
img = imread("colesterol-1.tif")
```

```python
imshow(img)
```

```python
fourier= fft2(img)
imshow(log(abs(fftshift(fourier))))
colorbar()
```

```python
high_freq = fftshift(fourier)
ic=int(fourier.shape[0])/2
jc=int(fourier.shape[1])/2
a=5E-3
for i in range(fourier.shape[0]):
    for j in range(fourier.shape[1]):
        high_freq[i,j]*=(1-exp(-a*((i-ic)**2 + (j-jc)**2)))
```

```python
im_high= ifft2(fftshift(high_freq))
imshow(abs(im_high))
colorbar()
```

##Normalización de la Imágen

```python
im_high[abs(im_high)<25]=0
im_high=abs(im_high)/abs(abs(im_high).max())
```

```python
imshow(im_high)
colorbar()
```

##Modelamiento


```python
def model(theta,r0,r1,m,phi):
    return r0+r1*sin(m*theta+phi)
def loglikelihood(im, grid):
    chi_squared = 0.5*sum((grid-im)**2)
    return -chi_squared
```

```python
r0_walk = empty((0))
r1_walk = empty((0))
phi_walk = empty((0))
m_walk = empty((0))
a_walk = empty((0))
b_walk = empty((0))
l_walk = empty((0))

r0_walk = append(r0_walk, 90)
r1_walk = append(r1_walk, random.random())
m_walk = append(m_walk, 1)
a_walk = append(a_walk, 175)
b_walk = append(b_walk, 110)
phi_walk = append(phi_walk, random.random())

theta = linspace(0,2*pi,1000)
grid = np.zeros((244,357))
for t in theta:
    j=int(b_walk[0] + sin(t)*model(t,r0_walk[0],r1_walk[0],m_walk[0],phi_walk[0]))
    i=int(a_walk[0] + cos(t)*model(t,r0_walk[0],r1_walk[0],m_walk[0],phi_walk[0]))
    grid[j,i]=1

print grid.max()
    
print r0_walk
print r1_walk
print phi_walk
print m_walk
print a_walk
print b_walk
```

```python
imshow(grid)
colorbar()
grid.shape
```

```python
l_walk = append(l_walk,loglikelihood(grid,im_high))
```

```python
n_iterations = 10000
for i in range(n_iterations):
    a_prime = np.random.normal(a_walk[i], 0.1) 
    b_prime = np.random.normal(b_walk[i], 0.1)
    m=m_walk[i]
    if(m_walk[i]==0 and numpy.random.random()>0.5):
        m=1
    m_prime = np.random.poisson(m, 1)[0]
    r0_prime = abs(np.random.normal(r0_walk[i], 1.0))
    r1_prime = abs(np.random.normal(r1_walk[i], 0.1))
    phi_prime = np.random.normal(phi_walk[i], 0.1)

    grid_prime=np.zeros(((244,357)))
    posj=int_(b_walk[i] + sin(theta)*model(theta,r0_prime,r1_prime,m_prime,phi_prime))
    posi=int_(a_walk[i] + cos(theta)*model(theta,r0_prime,r1_prime,m_prime,phi_prime))
    grid_prime[posj, posi]=1
    
    l_prime = loglikelihood(im_high, grid_prime)
    l_init = l_walk[i]

    a = l_prime - l_init
    if(a>=0):
        a_walk  = append(a_walk, a_prime)
        b_walk  = append(b_walk, b_prime)
        m_walk = append(m_walk, m_prime)
        r0_walk = append(r0_walk, r0_prime)
        r1_walk = append(r1_walk, r1_prime)
        phi_walk = append(phi_walk, phi_prime)
        l_walk = append(l_walk, l_prime)
    else:
        beta = random.random()
        if(beta<=exp(a)):
            a_walk  = append(a_walk,a_prime)
            b_walk  = append(b_walk,b_prime)
            m_walk = append(m_walk,m_prime)
            r0_walk = append(r0_walk,r0_prime)
            r1_walk = append(r1_walk,r1_prime)
            phi_walk = append(phi_walk,phi_prime)
            l_walk = append(l_walk, l_prime)
        else:
            a_walk  = append(a_walk,a_walk[i])
            b_walk  = append(b_walk,b_walk[i])
            m_walk = append(m_walk,m_walk[i])
            r0_walk = append(r0_walk,r0_walk[i])
            r1_walk = append(r1_walk,r1_walk[i])
            phi_walk = append(phi_walk,phi_walk[i])
            l_walk = append(l_walk, l_walk[i])
```

```python
fig, ax = plt.subplots(2, 3, figsize=(15, 5))
ax[0,0].scatter(a_walk,b_walk)
ax[0,1].plot(r0_walk)
ax[0,2].plot(r1_walk)
ax[1,0].plot(m_walk)
ax[1,1].plot(phi_walk)
ax[1,2].plot(l_walk)
```

```python
max_likelihood_id = argmax(l_walk)
best_a = a_walk[max_likelihood_id]
best_b = b_walk[max_likelihood_id]
best_m = m_walk[max_likelihood_id]
best_phi = phi_walk[max_likelihood_id]
best_r0 = r0_walk[max_likelihood_id]
best_r1 = r1_walk[max_likelihood_id]
print best_a
print best_b
print best_m
print best_r0
print best_r1
```

##Resultado Final
A continuación se muestra la imágen original junto con el rezultado hallado. A simplemente vista se observa una buena correlación. Además si observamos el comportamiento de l_walk. Observamos un comportamiento como el esperado.

```python
plot(best_a + cos(theta)*model(theta,best_r0,best_r1,best_m,best_phi), best_b + sin(theta)*model(theta,best_r0,best_r1,best_m,best_phi))
imshow(img, cmap = cm.Greys_r)
```
